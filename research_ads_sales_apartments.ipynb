## Предобработка данных

В пропущенных значениях столбца balcony поставлю 0, скорее всего это значение пропущено или, возможно, балкона просто нет. Далее заполню пропуски в столбце living_area средним значением для квартир примерно с такой площадью, предварительно сделаю столбец где разобью все объекты на группы по общей площади, аналогично поступлю с пропусками в kitchen_area. 

Для начала создам копию датафрэйма, и все изменения буду проводить в ней.

```python
df_copy = df.copy()
```

```python
df_copy['balcony'].fillna(value = 0, inplace=True)
```

Меня смущает минимальное значение стоимости объекта

```python
df_copy.query('last_price == 12190')
```

Получается этот объект находится в 9176 м от центра СПб. Т.е. он в СПБ в 9-ти км от центра с площадью 109 кв метров, здесь явно ошиблись в цене и не добавили 3 нуля. Быстро это поправлю и продолжу.

```python
df_copy['last_price'].loc[df_copy['last_price'] == 12190] = df_copy['last_price'].loc[df_copy['last_price'] == 12190] * 1000
```

теперь проверю колонки с площадью на предмет нет ли объектов в которых жилая площадь больше общей, площадь кухни больше общей, сумма жилой и кухни больше общей.

```python
len(df_copy.query('living_area > total_area'))
```

```python
len(df_copy.query('kitchen_area > total_area'))
```

```python
df_copy.query('(kitchen_area + living_area) > total_area')
```

Итак таких объектов довольно много. Присвою значения каждой из площадей (жилой, кухонной и общей) медианные значения для объектов с таким же количеством комнат.

```python
def area(a,b,c):
    return df_copy[c].where(a, b)
```

```python
rooms = list(df_copy.query('(kitchen_area + living_area) > total_area')['rooms'].unique())
for n_room in rooms:
    tot_a = 0
    area_values = ['kitchen_area','living_area']
    f_n = df_copy.query('(kitchen_area + living_area) > total_area and rooms == @n_room')
    for area_value in area_values:
        med_a = f_n[area_value].median()
        tot_a += med_a
        #f_n[area_value] = med_a
        df_copy[area_value].loc[((df_copy['living_area'] + df_copy['kitchen_area']) > df_copy['total_area']) & (df_copy['rooms'] == n_room)] = med_a
        df_copy['total_area'].loc[((df_copy['living_area'] + df_copy['kitchen_area']) > df_copy['total_area']) & (df_copy['rooms'] == n_room)] = tot_a
    
    df_copy['total_area'].loc[((df_copy['living_area'] + df_copy['kitchen_area']) > df_copy['total_area']) & (df_copy['rooms'] == n_room)] = tot_a
    
    
    
   
```

Затем построю гистограмму по значениям общей площади, чтобы понять распределение данных

```python
#df_copy.hist('total_area', bins = 20, range = (0, 300))
plt.hist(df_copy['total_area'], bins = 20, range = (0, 300))
plt.title('гистограмма распределения площади объектов')
plt.xlabel('площадь')
plt.ylabel('количество объектов')
plt.show()
```

<div class="alert alert-info">
Сделал подписи к графику</div>

По гистограмме видно, что большинство объектов имеет площадь от 30 до 60 кв метров, менее всего объектов площадью больше 190 кв метров. Распределение данных на гистограмме Пуассоновское.  

Добавлю столбец с номерами категорий в зависимости от плащади.

```python
df_copy['cat_total_area'] = 0
```

```python
n, m = 0, 13
for num_cat in range(20):
    step = 13
    df_copy['cat_total_area'].loc[(df_copy['total_area'] <= m) & (df_copy['total_area'] > n)] = num_cat
    n += step
    m += step
    
df_copy['cat_total_area'].loc[df_copy['total_area'] > (13*21)] = 21
```

```python
mean_living_area = []
mean_kitchen_area = []
for i in range(max(df_copy['cat_total_area'])+1):
    mean_living_area.append(df_copy.query(f'cat_total_area == {i}')['living_area'].mean())
    mean_kitchen_area.append(df_copy.query(f'cat_total_area == {i}')['kitchen_area'].mean())
```

```python
for cat in range(len(mean_living_area)):
    df_copy['living_area'].loc[(df_copy['living_area'].isna()) & (df_copy['cat_total_area'] == cat)] = mean_living_area[cat] 
    df_copy['kitchen_area'].loc[(df_copy['kitchen_area'].isna()) & (df_copy['cat_total_area'] == cat)] = mean_kitchen_area[cat]
```

Далее исследую столбец ceiling_height, вычислю среднее, предварительно отбросив выбросы,  и заполню им пропуски.

```python
#df_copy.hist('ceiling_height', bins = 20, range=(0,10))
plt.hist(df_copy['total_area'], bins = 20, range = (0, 300))
plt.title('гистограмма распределения высоты потолков объектов')
plt.xlabel('высота потолков')
plt.ylabel('количество объектов')
plt.show()
```

<div class="alert alert-info">
Сделал подписи к графику</div>

```python
mean_ceiling_height = df_copy.query('ceiling_height < 10 or ceiling_height > 2')['ceiling_height'].mean()
df_copy['ceiling_height'].fillna(value = mean_ceiling_height, inplace=True)
```

Теперь заполню пропуски в floors_total присвою им значения floor для этого объекта.

```python
floors_total_na = df_copy.loc[df_copy['floors_total'].isna()]
for line in floors_total_na.index:
    #print(floors_total_na)
    df_copy['floors_total'][line] = df_copy['floor'][line] 
```

Далее заполню пропуски в is_apartment присвоив им значение 0, остальные заменю на 1.

```python
df_copy['is_apartment'].fillna(0, inplace = True)
```

```python
df_copy.query('is_apartment != 0')['is_apartment'] = 1
```

Теперь заполню какие смогу пропуски в locality_name, вычислю максимальное значение от центра СПб до его границы и приму, что все объекты имеющие расстояние от центра до этого значения, находятся в СПб.

```python
locality_name_na = df_copy.loc[df_copy['locality_name'].isna()]
```

```python
spb_area = df_copy.query('locality_name == "Санкт-Петербург"')['cityCenters_nearest'].max()
```

```python
for num in locality_name_na.index:
    if df_copy['cityCenters_nearest'][num] <= spb_area:
        df_copy['locality_name'][num] = 'Санкт-Петербург'
```

Произведу замену типа данных в следующих столбцах: 
1) is_apartment на булевский тип т.к. в данном столбце возможны только 2 значения;
2) balcony на целочисленный т.к. у объекта не может быть 2,5 балкона;
3) floors_total на целочисленный т.к. у объекта может быть только целое число этажей.

Также заменю тип столбца first_day_exposition на datetime

Не следует производить замену типа в следующих столбцах: parks_around3000, parks_nearest, ponds_around3000, ponds_nearest т. к. в них присутствуют значения NaN, что, скорее всего, означает отсутствия объектов с названием столбца, и NaN соответствует типу Float, что дает нам возможность производить над данными в этих столбцах математические действия.  По тем же причинам не изменяю тип данных столбца days_exposition.

```python
df_copy['is_apartment'] = df_copy['is_apartment'].astype(bool)
```

```python
df_copy['balcony'] = df_copy['balcony'].astype(int)
```

```python
df_copy['floors_total'] = df_copy['floors_total'].astype(int)
```

```python
df_copy.info()
```

Вывод: на данном этапе я заполнил пропуски и совсем немного поработал с выбросами. Теперь можно приступить к более глубокому изучению датасета.

```python
df_copy['kitchen_area/total_area'] = df_copy['kitchen_area']/df_copy['total_area']
```
